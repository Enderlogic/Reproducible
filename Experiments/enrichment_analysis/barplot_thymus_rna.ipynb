```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# group_name_map = {
#     0: 'WM',
#     1: 'L1',
#     2: 'L2',
#     3: 'L3',
#     4: 'L4',
#     5: 'L5',
#     6: 'L6'
# }

# Assuming df1['original_clusters'] has values from 0 to 6
out_path = '/maiziezhou_lab2/yunfei/Projects/Spinal_MERFISH/GSVA/data_in/yangv2/Mouse Thymus results (RNA+Protein)/out/rna/m2'
os.makedirs(out_path, exist_ok=True)
# Transpose df2 to make indexing easier (optional but helps)
# Now rows = cells, columns = pathways
df2_T = df_m2.T

# Store results
log2fc_dict = {}

for group in groups:
    # Identify cell indices
    # in_group = group_df[group_df['protein_group'] == group & (group_df['protein_group'] != 'none')].index
    in_group = group_df[(group_df['transcriptomic_group'] == group) & (group_df['transcriptomic_group'] != 'none')].index
    rest = group_df[(group_df['transcriptomic_group'] != group) & (group_df['transcriptomic_group'] != 'none')].index

    # Calculate means
    mean_in_group = df2_T.loc[in_group].mean()
    mean_rest = df2_T.loc[rest].mean()

    # Compute log2 fold change
    epsilon = 1e-9
    log2fc = np.log2((mean_in_group + epsilon) / (mean_rest + epsilon))

    # Save full values if needed
    log2fc_dict[group] = log2fc

    # Select top 10 by absolute value
    top10 = log2fc.sort_values(ascending=False).head(10)

    # # Plot
    # plt.figure(figsize=(10, 5))
    # sns.barplot(x=top10.values, y=top10.index, palette="vlag")
    # plt.title(group)
    # plt.xlabel('log2 Fold Change')
    # plt.ylabel('')
    # plt.tight_layout()
    # # plt.savefig(os.path.join(out_path, str(group) + 'top10.pdf'))
    # plt.show()
    # Filter to positive + not null values only
    filtered = top10[(top10 > 0) & (~top10.isna())]

    # Plot only if there's something to show
    if not filtered.empty:
        bar_height = 0.5  # tweak this if labels are overlapping
        n_bars = len(filtered)
        height = max(4, n_bars * bar_height)  # minimum height of 4

        plt.figure(figsize=(14, height))
        sns.barplot(x=filtered.values, y=filtered.index, palette="vlag")
        plt.title(group)
        plt.xlabel('log2 Fold Change')
        plt.ylabel('')
        plt.tight_layout()
        
        plt.savefig(os.path.join(out_path, str(group) + 'top.pdf'), bbox_inches='tight')
        plt.show()
        print(filtered.values)
```

```python
import warnings
warnings.filterwarnings("ignore")
import random
import os
import scanpy as sc
import cellcharter as cc
import pandas
import sys
sys.path.append('/home/makx/Reproducible-main')
from Methods.SpaMV.utils import pca, clr_normalize_each_cell
from sklearn.metrics import adjusted_rand_score
import scvi
import matplotlib as mpl
import squidpy as sq
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import anndata as ad
from matplotlib.colors import ListedColormap
import h5py
import pickle
import anndata
from lightning.pytorch import seed_everything




# load simulated dataset
for dataset in ['1_Simulation', '2_Simulation', '3_Simulation']:
    data_rna = sc.read_h5ad('../Dataset/' + dataset + '/adata_RNA.h5ad')
    data_pro = sc.read_h5ad('../Dataset/' + dataset + '/adata_ADT.h5ad')
    data_rna.layers["counts"] = data_rna.X.copy()
    data_pro.layers["counts"] = data_pro.X.copy()

    # RNA
    sc.pp.filter_genes(data_rna, min_cells=10)
    sc.pp.normalize_total(data_rna, target_sum=1e4)
    sc.pp.log1p(data_rna)
    sc.pp.scale(data_rna)
    data_rna.obsm['feat'] = pca(data_rna, n_comps=50)

    # Protein
    data_pro = clr_normalize_each_cell(data_pro)
    sc.pp.scale(data_pro)
    data_pro.obsm['feat'] = pca(data_pro, n_comps=50)

    adata = sc.concat([data_rna, data_pro], axis=1, merge='same')
    adata.obs['sample'] = dataset
    data_rna.obs['sample'] = dataset
    data_pro.obs['sample'] = dataset
    data_rna.obs['sample'] = pd.Categorical(data_rna.obs['sample'])
    data_pro.obs['sample'] = pd.Categorical(data_pro.obs['sample'])
    adata.obs['sample'] = pd.Categorical(adata.obs['sample'])
    adata.obsm['spatial'] = data_rna.obsm['spatial']
    adata.uns['spatial'] = {s: {} for s in adata.obs['sample'].unique()}
    adata.obs['cluster'] = data_rna.obs['cluster'].copy()
    


    model_omics1 = scvi.model.SCVI.setup_anndata(
    data_rna,
    layer="counts",
    batch_key='sample',)
    model_omics1 = scvi.model.SCVI(data_rna)
    model_omics1.train(early_stopping=True, enable_progress_bar=True)
    data_rna.obsm['X_latent'] = model_omics1.get_latent_representation()

    model_omics2 = scvi.model.SCVI.setup_anndata(
    data_pro,
    layer="counts",
    batch_key='sample',)
    model_omics2 = scvi.model.SCVI(data_pro)
    model_omics2.train(early_stopping=True, enable_progress_bar=True)
    data_pro.obsm['X_latent'] = model_omics2.get_latent_representation()

    adata.obsm['X_latent'] = np.concatenate([data_rna.obsm['X_latent'], data_pro.obsm['X_latent']], axis=1)
    sq.gr.spatial_neighbors(adata, library_key='sample', coord_type='grid', n_neighs=4, n_rings=1)
    cc.gr.aggregate_neighbors(adata, n_layers=4, use_rep='X_latent')
    gmm = cc.tl.Cluster(
    n_clusters=10,)
    gmm.fit(adata, use_rep='X_cellcharter')
    adata.obs['X_cellcharter'] = gmm.predict(adata, use_rep='X_cellcharter')
    adata.obs['X_cellcharter'] = adata.obs['X_cellcharter'].cat.codes + 1
    adata.obs['X_cellcharter'] = adata.obs['X_cellcharter'].astype('category')
    custom_palette = {
    1: '#1f77b4',  # 蓝色
    2: '#ff7f0e',  # 橙色
    3: '#2ca02c',  # 绿色
    4: '#d62728',  # 红色
    5: '#9467bd',  # 紫色
    6: '#8c564b',  # 棕色
    7: '#e377c2',  # 粉色
    8: '#7f7f7f',  # 灰色
    9: '#bcbd22',  # 黄绿色
    10: '#17becf',  # 青色
    }
    fig, ax_list = plt.subplots(1, 2, figsize=(8, 4))
    sc.pp.neighbors(adata, use_rep='X_cellcharter', n_neighbors=30)
    sc.tl.umap(adata)
    sc.pl.umap(adata, color='X_cellcharter', ax=ax_list[0], title='CellCharter\n' + dataset, s=60, show=False,palette=custom_palette)
    sc.pl.embedding(adata, basis='spatial', color='X_cellcharter', ax=ax_list[1], title='CellCharter\n' + 'ARI: {:.3f}'.format(adjusted_rand_score(adata.obs['X_cellcharter'], adata.obs['cluster'])), s=200, show=False,palette=custom_palette)

    plt.tight_layout(w_pad=0.3)
    result_folder = '../Results/' + dataset + '/'
    if not os.path.exists(result_folder):
        os.makedirs(result_folder, exist_ok=True)
    plt.savefig(result_folder + 'CellCharter.pdf')
    plt.show()
```
